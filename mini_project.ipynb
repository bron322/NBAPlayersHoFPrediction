```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
rookiedata = pd.read_csv('rookie_data.csv')
rookiedata.head()
```

```python
rookiedata['Hall of Fame Class'] = rookiedata['Hall of Fame Class'].apply(lambda x:0 if pd.isna(x) else 1)
rookiedata.head(40)
```

I cleaned the data by making the Hall of Fame Class to have either 0 and 1 data. 0 means not in Hall of Fame whereas 1 means in Hall of Fame.

```python
#removing outliers that has the lowest gameplays
Q1 = rookiedata['GP'].quantile(0.25)
Q3 = rookiedata['GP'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
rookiedata = rookiedata[rookiedata['GP'] >= lower_bound]
```

```python
corr_matrix = rookiedata.corr()
corr_with_eff = corr_matrix.loc['EFF']
corr_with_eff_sorted = corr_with_eff.sort_values(ascending=False)
print(corr_with_eff_sorted)
```

```python
# Import train_test_split from sklearn
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(rookiedata, test_size=0.2, random_state=0)
```

```python
# Import LinearRegression model from Scikit-Learn
from sklearn.linear_model import LinearRegression
# Import mean_squared_error from sklearn
from sklearn.metrics import mean_squared_error

def regression(data,target_var, predictor):
    train_x = train_data[[predictor]].values
    train_y = train_data[[target_var]].values
    test_x = test_data[[predictor]].values
    test_y = test_data[[target_var]].values
    reg = LinearRegression().fit(train_x,train_y)
    intercept = reg.intercept_
    coefficients = reg.coef_
    
    # Formula for the Regression line
    regline_x = train_x
    regline_y = reg.intercept_ + reg.coef_ * train_x

    # Plot the Linear Regression line
    f, axes = plt.subplots(1, 1, figsize=(16, 8))
    plt.scatter(train_x, train_y)
    plt.plot(regline_x, regline_y, 'r-', linewidth = 3)
    plt.show()
    
    # Predict SalePrice values corresponding to GrLivArea
    y_test_pred = reg.predict(test_x)
    
    # Plot the Predictions on a Scatterplot
    f = plt.figure(figsize=(16, 8))
    plt.scatter(test_x, test_y, color = "green")
    plt.scatter(test_x, y_test_pred, color = "red")
    plt.show()
    
    # Explained Variance in simply the "Score"
    print("Explained Variance (R^2) \t:", reg.score(train_x, train_y))
    
    # Compute MSE on the test set
    print("Mean Squared Error (MSE) \t:", mean_squared_error(test_y, y_test_pred))
    print("Intercept:" , intercept)
    print("Coefficients: " , coefficients)
    return intercept, coefficients
```

```python
# predict Efficiency using PTS
intercept, coefficients = regression(rookiedata, "EFF", "PTS")
```

```python
# predict Efficiency using PTS
intercept, coefficients = regression(rookiedata, "EFF", "FGM")
```

```python
# predict Efficiency using MIN
intercept, coefficients = regression(rookiedata, "EFF", "MIN")
```

```python
# Extract Response and Predictors
y = pd.DataFrame(rookiedata["EFF"])
X = pd.DataFrame(rookiedata[["PTS", "FGM", "MIN"]])

# Split the Dataset into Train and Test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

# Check the sample sizes
print("Train Set :", y_train.shape, X_train.shape)
print("Test Set  :", y_test.shape, X_test.shape)
```

```python
# Summary Statistics for Predictors
X_train.describe()
```

```python
# Draw the distribution of Response
f, axes = plt.subplots(1, 3, figsize=(24, 6))
sb.boxplot(data = y_train, orient = "h", ax = axes[0])
sb.histplot(data = y_train, ax = axes[1])
sb.violinplot(data = y_train, orient = "h", ax = axes[2])
```

```python
# Draw the distributions of all Predictors
f, axes = plt.subplots(3, 3, figsize=(18, 12))

count = 0
for var in X_train:
    sb.boxplot(data = X_train[var], orient = "h", ax = axes[count,0])
    sb.histplot(data = X_train[var], ax = axes[count,1])
    sb.violinplot(data = X_train[var], orient = "h", ax = axes[count,2])
    count += 1
```

```python
# Correlation between Response and the Predictors
trainDF = pd.concat([y_train, X_train], axis = 1).reindex(y_train.index)

f = plt.figure(figsize=(12, 8))
sb.heatmap(trainDF.corr(), vmin = -1, vmax = 1, annot = True, fmt = ".2f")
```

```python
# Relationship between Response and the Predictors
sb.pairplot(data = trainDF)
```

## We hypothesise a linear regression model:
EFF = a1 x PTS + a2 x FGM + a3 x MIN + b
      where a1 is PTS coefficient, a2 is FGM coefficient, a3 is MIN coefficient and b is the intercept of regression.

```python
# Linear Regression using Train Data
linreg = LinearRegression()         # create the linear regression object
linreg.fit(X_train, y_train)        # train the linear regression model

# Coefficients of the Linear Regression line
print('Intercept of Regression \t: b = ', linreg.intercept_)
print('Coefficients of Regression \t: a = ', linreg.coef_)
print()

# Print the Coefficients against Predictors
pd.DataFrame(list(zip(X_train.columns, linreg.coef_[0])), columns = ["Predictors", "Coefficients"])
```

```python
# Predict the Total values from Predictors 
y_train_pred = linreg.predict(X_train) 
y_test_pred = linreg.predict(X_test) 
 
# Plot the Predictions vs the True values 
f, axes = plt.subplots(1, 2, figsize=(24, 12)) 
axes[0].scatter(y_train, y_train_pred, color = "blue") 
axes[0].plot(y_train, y_train, 'w-', linewidth = 1) 
axes[0].set_xlabel("True values of the Response Variable (Train)") 
axes[0].set_ylabel("Predicted values of the Response Variable (Train)") 
axes[1].scatter(y_test, y_test_pred, color = "green") 
axes[1].plot(y_test, y_test, 'w-', linewidth = 1) 
axes[1].set_xlabel("True values of the Response Variable (Test)") 
axes[1].set_ylabel("Predicted values of the Response Variable (Test)") 
plt.show()

# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model \tTrain Dataset")
print("Explained Variance (R^2) \t:", linreg.score(X_train, y_train))
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_train, y_train_pred))
print()

# Check the Goodness of Fit (on Test Data)
print("Goodness of Fit of Model \tTest Dataset")
print("Explained Variance (R^2) \t:", linreg.score(X_test, y_test))
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_test, y_test_pred))
print()
```

## As we can see, when we were doing uni-variate linear regression, the R^2 value was just 0.845, 0.843 , 0.829 for PTS, FGM, MIN as predictors respectively. When we use multi-variate linear regression, we have R^2 value to be 0.884 which is higher than when we were doing uni-variate linear regression.
